```python
def generate_array_time(current_time,end_time,move,time):
    app=[]
    while current_time <= end_time:
       
        current_time += move
        if time:
            app.append(current_time.strftime("%H:%M:%S"))
        else:
            app.append(current_time.strftime("%Y-%m-%d"))
    return app
start_date=datetime.date(2024,4,14)

end_date=datetime.date(2024,4,29)

move=datetime.timedelta(days=1)
dates=generate_array_time(start_date,end_date,move,False)
dates
```
